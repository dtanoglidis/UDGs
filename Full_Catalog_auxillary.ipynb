```python
RAh, RAm, RAs, DECd,DECm, DECs = np.loadtxt("Reflex_RADEC.dat",  unpack=3, usecols=[1,2,3,4,5,6])

RA_reflex = RAh*15.0 + RAm/4.0 + RAs/240.0  # Right ascension in degrees 

# Get signs 
sig = np.sign(DECd)
sig[32]=sig[104]=sig[245]=sig[288]=sig[436]=1
sig[174]=sig[175]=sig[272]=sig[289]=-1

DEC_reflex = sig*(abs(DECd) + DECm/60.0 + DECs/3600.0)
```

```python
plt.figure(figsize = (8.0,8.0))
smap = DESSkymap()
#smap.scatter(ra_neg, dec_neg, c='r', s=0.05,latlon=True, label="non-LSB objects")
smap.scatter(ra_posit, dec_posit, c='r', s=0.5,latlon=True, label="LSB objects")
#smap.scatter(ra, dec, c='r', s=0.0005,latlon=True, label="LSB objects")
smap.scatter(RA_reflex, DEC_reflex, c='darkblue', s=3., latlon=True, label="reflex clusters")
smap.draw_des(color='darkblue',label='DES footprint')

smap.scatter(18.5, -31.5, c='yellow', s=4., latlon=True, label="cluster")


plt.xlabel("RA", fontsize=15)
plt.ylabel("DEC", fontsize=15)




plt.legend(frameon=True, loc='upper left', fontsize=10)

plt.savefig('Positives.pdf')
plt.show()
```
