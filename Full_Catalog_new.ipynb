```python
# Load the model from the file 
SVM_class = joblib.load('Classifier.pkl') 
```

```python
# Define the ellipticity now
ellipticity = 1.0 - B_IMAGE/A_IMAGE

# ===========================================================================================
# Define the colors
col_g_r = MAG_AUTO_G - MAG_AUTO_R
col_g_i = MAG_AUTO_G - MAG_AUTO_I
col_i_r = MAG_AUTO_G - MAG_AUTO_R
```

```python
# Define the train data matrix of features
X_feat = np.zeros([len(ra),18])
X_feat[:,0] = MAG_AUTO_G;X_feat[:,1] = MAG_AUTO_R;X_feat[:,2] = MAG_AUTO_I
X_feat[:,3] = col_g_r;X_feat[:,4] = col_g_i;X_feat[:,5] = col_i_r
X_feat[:,6] = FLUX_RADIUS_G;X_feat[:,7] = FLUX_RADIUS_R;X_feat[:,8] = FLUX_RADIUS_I
X_feat[:,9] = MU_EFF_MODEL_G;X_feat[:,10] = MU_EFF_MODEL_R;X_feat[:,11] = MU_EFF_MODEL_I
X_feat[:,12] = MU_MAX_G; X_feat[:,13] =  MU_MAX_R;X_feat[:,14] =  MU_MAX_I
X_feat[:,15] = ellipticity;X_feat[:,16] = A_IMAGE; X_feat[:,17] = B_IMAGE
```

```python
from sklearn import preprocessing
scaler = joblib.load('Scaler.pkl') 
X_feat = scaler.transform(X_feat)

y_pred = SVM_class.predict(X_feat)
```

```python
RA_pos = ra[y_pred==1.]
DEC_pos = dec[y_pred==1.]
```

### Auto-correlations

```python
import treecorr
import fitsio
```

```python
# Create catalog. For now, we choose our catalog to be the objects classified as real LSB objects 
cat1 = treecorr.Catalog(ra=RA_pos, dec=DEC_pos, ra_units='deg', dec_units='deg')
```

```python
dd_1 = treecorr.NNCorrelation(min_sep=0.01, max_sep=20, bin_size=0.1, sep_units='degrees')
dd_1.process(cat1)
```

```python
leng_1 = len(RA_pos)
ra_min = np.min(cat1.ra)
ra_max = np.max(cat1.ra)
dec_min = np.min(cat1.dec)
dec_max = np.max(cat1.dec)

rand_ra_1 = np.random.uniform(ra_min, ra_max, leng_1) #Random RA
rand_sindec_1 = np.random.uniform(np.sin(dec_min), np.sin(dec_max), leng_1)
rand_dec_1 = np.arcsin(rand_sindec_1)
```

```python
rand_1 = treecorr.Catalog(ra=rand_ra_1, dec=rand_dec_1, ra_units='radians', dec_units='radians')
rr_1 = treecorr.NNCorrelation(min_sep=0.01, max_sep=20, bin_size=0.1, sep_units='degrees')
rr_1.process(rand_1)
```

```python
xi, varxi_1 = dd_1.calculateXi(rr_1)
r = numpy.exp(dd_1.meanlogr)
sig = numpy.sqrt(varxi_1)

plt.plot(r, xi, color='blue')
plt.plot(r, -xi, color='blue', ls=':')
plt.errorbar(r[xi>0], xi[xi>0], yerr=sig[xi>0], color='blue', lw=0.1, ls='')
plt.errorbar(r[xi<0], -xi[xi<0], yerr=sig[xi<0], color='blue', lw=0.1, ls='')
leg = plt.errorbar(-r, xi, yerr=sig, color='blue')

plt.xscale('log')
plt.yscale('log', nonposy='clip')
plt.xlabel(r'$\theta$ (degrees)')
plt.ylabel(r'$w(\theta)$')

#plt.legend([leg], [r'$w(\theta)$'], loc='upper right')
plt.xlim([0.01,20])
plt.savefig('Autocorr.pdf')
plt.show()
```

### Cross correlations

```python
# Catalog of the objects classified as LSB
cat_LSB = treecorr.Catalog(ra=RA_pos, dec=DEC_pos, ra_units='deg', dec_units='deg') 
# Catalog of the clusters
cat_clust = treecorr.Catalog(ra=RA_reflex, dec=DEC_reflex , ra_units='deg', dec_units='deg')
```

```python
NN_corr = treecorr.NNCorrelation(min_sep=0.01, max_sep=20, bin_size=0.1, sep_units='degrees')
NN_corr.process(cat_LSB,cat_clust) 
```

```python
# Random galaxies
leng_1 = len(RA_pos)
ra_min = np.min(cat_LSB.ra)
ra_max = np.max(cat_LSB.ra)
dec_min = np.min(cat_LSB.dec)
dec_max = np.max(cat_LSB.dec)

rand_ra_1 = np.random.uniform(ra_min, ra_max, leng_1) #Random RA
rand_sindec_1 = np.random.uniform(np.sin(dec_min), np.sin(dec_max), leng_1)
rand_dec_1 = np.arcsin(rand_sindec_1)

# Random clusters
leng_2 = len(RA_reflex)

rand_ra_2 = np.random.uniform(ra_min, ra_max, leng_2) #Random RA
rand_sindec_2 = np.random.uniform(np.sin(dec_min), np.sin(dec_max), leng_2)
rand_dec_2 = np.arcsin(rand_sindec_2)
```

```python
# Random catalogs
rand_gal = treecorr.Catalog(ra=rand_ra_1, dec=rand_dec_1, ra_units='radians', dec_units='radians')
rand_clust = treecorr.Catalog(ra=rand_ra_2, dec=rand_dec_2, ra_units='radians', dec_units='radians')

# Process them
rr = treecorr.NNCorrelation(min_sep=0.01, max_sep=20, bin_size=0.1, sep_units='degrees')
rr.process(rand_gal, rand_clust)
```

```python
# And cross-catalogues
# random - data
rd = treecorr.NNCorrelation(min_sep=0.01, max_sep=20, bin_size=0.1, sep_units='degrees')
rd.process(cat_LSB, rand_clust)
# data - random
dr = treecorr.NNCorrelation(min_sep=0.01, max_sep=20, bin_size=0.1, sep_units='degrees')
dr.process(cat_clust, rand_gal)
```

```python
xi, varxi = NN_corr.calculateXi(rr, dr, rd)
```
